```
인공신경망이란

인공신경망 (Artificial Neural Networks ANNs 이란 개념은 1943 년에 McCulloch, Warren S., and Walter Pitts. 가 발표한 A logical
calculus of the ideas immanent in nervous activity 이란 논문에서 최초로 제안되었다 McCulloch 과 Pitts 는 인간의 신경 구조를 
복잡한 스위치들이 연결된 네트워크 로 표현할 수 있다고 제안하였다.

전기신호를 받으면 각각의 가중치를 곱하고 특정값을 더한후 축색돌기를 통해 이동한다. 그값이 어떤 값 이산이라면 신경말단에서 다음뉴런으로 넘어가고 그이하이면 넘어가지 않는다


MLP 는 파라미터가 개수가 많아지면서 적절한 Weight 와 Bias 를 학습하는 것이 매우 어려웠는데 ,
McClelland, James L., David E. Rumelhart , and Geoffrey E. Hinton 은 Backpropagation
Algorithm 을 제안해서 이 문제를 해결하였다 .


Backpropagation Algorithm 오류 역전파 알고리즘 은 정방 (Feedforward) 연산 이후 ,
에러 - 예측 값과 트루 값과의 오차 를 후방 으로 다시 보내 줌으로써 , 많은
노드를 가진 MLP 라도 최적의 Weight 와 Bias 를 학습할 수 있도록 한다
```

```
Relu 

기존의 인공신경망은 위에서 언급한 over fitting 과 gradient vanishing(exploding)
problem 이 치명적인 결함으로 지적되어 왔었고 이를 해결하기 위해 여러
알고리즘이 발전했는데 , 그중에 하나가 ReLu funciton 이다

ReLu (Rectified Linear Unit) function 는 0 이하는 0, 0 이상은 x 값을 가지는 함수로
앞서 언급한 두 문제를 완화시키는데 효과적인 역할을 한다
```
